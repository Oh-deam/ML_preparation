1. R&D Spend - сколько потратила компания на исследования и разработки
2. Admission - сколько потратила на администрирование, например, на оплату труда
3. Marketing Spend - сколько было потрачено на маркетинг

5. Profit - прибыль

```python
df = pd.read_csv("data/50_Startups.csv")
df.head()
```

```python
df["State"].unique()
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

Закодируем колонку State

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False)
c = encoder.fit_transform(df[['State']])

encoded_cols = encoder.get_feature_names_out(['State'])
```

```python
df_encoded = pd.DataFrame(encoded, columns=encoded_cols)
df_encoded.head()
```

```python
df = pd.concat([df, df_encoded], axis=1)
```

```python
df.head()
```

Делаем множественную линейную регрессию

```python
X = df.drop(["State", "Profit"], axis=1)
y = df.loc[:, 'Profit']
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()

regressor.fit(X_train, y_train)
```

Concatenate - это функция NumPy, которая позволяет объединять вертикально или горизонтально, два вектора или даже массивы.

```python
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```
